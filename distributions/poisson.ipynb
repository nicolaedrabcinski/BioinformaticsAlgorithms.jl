```julia
import Pkg

Pkg.add("Distributions")
Pkg.add("Plots")
```

```julia
using Distributions
using Plots
```

```julia
# Define parameters
λ = 1000.0
```

```julia
# Create Poisson distribution
dist = Poisson(λ)
```

```julia
# Generate random samples
num_samples = 1000
random_samples = rand(dist, num_samples)
```

```julia
# Plotting histogram
histogram(random_samples, bins=50, xlabel="k", ylabel="Frequency", 
          title="Histogram of Poisson Distribution Samples (λ = $λ)", legend=false)
```
